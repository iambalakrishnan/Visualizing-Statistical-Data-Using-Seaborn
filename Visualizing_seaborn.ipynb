```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

%matplotlib inline
```

```python
athlete_data = pd.read_csv('athletes.csv')

country_data = pd.read_csv('countries.csv')

tips_data = sns.load_dataset('tips')
```

```python
athlete_data.head()
```

```python
country_data.head()
```

```python
tips_data.head()
```

```python
#Relationship between population and GDP

#Scatterplot 
sns.relplot(x='population', y='gdp_per_capita', data = country_data)
plt.show()
```

```python
#Correlation between weight and height among the athlete

#sns.relplot(athlete_data['height'], athlete_data['weight'])
sns.relplot(x = 'height', y='weight', data = athlete_data)
plt.show()
```

```python
#Height and weight correlation based on sex

sns.relplot(x = 'height', y='weight', hue='sex', data = athlete_data)
plt.show()
```

```python
sns.relplot(x = 'height', y='weight', hue='sport',data = athlete_data)
plt.show()
```

<h3>Plotting categorical data</h3>

<h4>connection between tip amount and day of week</h4>

```python
sns.catplot(x = 'day', y = 'tip', data = tips_data)
plt.show()
```

Swarm plot - prevents the points from overlapping and adjust the points along the categorical access.

```python
sns.catplot(x = 'day', y = 'tip', kind = 'swarm', data = tips_data)
plt.show()
```

Who typically  gets the heigher tips - males or females?

```python
sns.catplot(x = 'day', y = 'tip', hue='sex', kind = 'swarm', data = tips_data)
plt.show()
```

Boxplot

```python
sns.catplot(x = 'day', y = 'tip',  kind = 'box', data = tips_data)
plt.show()
```

```python
sns.boxplot(x='day',y='tip', data=tips_data)
```

```python
sns.catplot(x = 'day', y = 'tip', hue='sex', kind = 'box', data = tips_data)
plt.show()
```

```python
sns.catplot(x = 'day', y = 'tip', hue='sex',  kind = 'violin', height=7, aspect=2,data = tips_data)
plt.show()
```

What are the types of the sports and what does their participation look like?

```python
sns.countplot(x = athlete_data['sport'], order = athlete_data['sport'].value_counts().index)
plt.show()
```

<h4>Visualizing Distribution of Datasets</h4><br>

What does the total bill look like in the restaurant?


```python
sns.distplot(tips_data['total_bill'], kde=False)
```

KDE-Kernal Density Estimate

```python
sns.kdeplot(tips_data['total_bill'],shade=True)
```

What is the relationship between total_bill and tip?<br>

Bivaiate Plot - 2D KDE plot

```python
sns.jointplot(x = tips_data['total_bill'], y = tips_data['tip'], kind = 'kde', shade=True)
```
